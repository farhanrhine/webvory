```python
import pandas as pd 
```

```python
df = pd.read_csv('orders.csv')
df.head()
```

```python
# 1. Get unique values for 'status'
unique_statuses = df['status'].unique()
status_count = df['status'].nunique()

# Display Results
print(f"Unique Statuses ({status_count}): {unique_statuses}")
```

```python

# 2. Get unique values for 'tracking_id'
unique_tracking = df['tracking_id'].unique()
tracking_count = df['tracking_id'].nunique()

# Display Results
print(f"Unique Tracking IDs ({tracking_count}): {unique_tracking}")
```
